```python
import pickle
from cntmosaic.dataloader.restru_loaders import MergedLoader, RawLoader
import pandas as pd
import numpy as np
import xarray as xr
from cntmosaic.dataloader import CoordToColumns
```

```python

with open('../datasets/data/polymod_germany.pkl', 'rb') as file:
    data = pickle.load(file)['contacts']
```

```python
data['binned'] = pd.cut(data.age_cnt, bins=range(0,86,5), right=False)
```

```python
with open('../datasets/data/polymod_germany.pkl', 'rb') as file:
    pop = pickle.load(file)['population']
```

```python
mapper  = CoordToColumns(age_part='age_part', age_grp_cnt='binned', id_var='id',  population_age='age', population_size='P')
```

```python
d = MergedLoader(data, mapper, pop[pop.age < 85])
```

```python
d.load()
```

```python
from cntmosaic.models.restr_BRCrefine import restr_BRCrefine
```

```python
model = restr_BRCrefine(d)
```

```python
model.params.grid_type = 'diff-age'
```

```python
model.run_inference_mcmc(0,1000,500,2)
```

```python
from cntmosaic.analysis import ModelSummariserMCMC, ModelVisualiser
```

```python
MSM = ModelSummariserMCMC(model)
```

```python
visual = ModelVisualiser(MSM)
```

```python
visual.plot_cint()['general']
```

```python
visual.plot_rate()
```
