```python
import numpyro
from numpyro.distributions import Normal

import jax
import jax.numpy as jnp
from jax import random
```

```python
rng_key = random.PRNGKey(0)
```

```python
with numpyro.plate('x_plate', 2, dim=-1):
  x = numpyro.sample('x', Normal(0, 1).expand([3,3]).to_event(2), rng_key=rng_key)
x.shape
```

```python
a = numpyro.sample('a', Normal(0, 1).expand([2]), rng_key=rng_key)
b = numpyro.sample('b', Normal(0, 1), rng_key=rng_key)
c = numpyro.sample('c', Normal(0, 1), rng_key=rng_key)

x = jnp.arange(5)
X, Y = jnp.ix_(x, x)

(a[:, None, None] + b * X + c * Y).shape
```
